```python
## chat completion api

response =openai.ChatCompletion.create(
    model = 'gpt-3.5-turbo',
    messages = [{'role' : 'system', 'content' : "Act as an AI assitant"},
    {'role' : 'user', 'content' : "Hello How are you?"}
    ]
)
response
```

```python
## chat completion api

response =openai.ChatCompletion.create(
    model = 'gpt-3.5-turbo',
    messages = [{'role' : 'system', 'content' : "Act as an AI assitant"},
    {'role' : 'user', 'content' : "Hello How are you?"},
    {'role' : 'user', 'content' : "My age is 33 and I am a Manager"},
    {'role': 'user', 'content' : 'What is my age'}
    ]
)
response
```
